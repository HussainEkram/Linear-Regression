```python
#1 What is Simple Linear Regression (SLR)? Explain its purpose.
'''
Simple Linear Regression (SLR) is a supervised machine learning technique used to model the relationship between one independent variable (X) and one dependent variable (Y) using a straight line.
The relationship is represented by the equation:

Y = mX + c

where:

Y ‚Üí Dependent variable (output or target)

X ‚Üí Independent variable (input or predictor)

m ‚Üí Slope of the line (shows how much Y changes when X changes)

c ‚Üí Intercept (value of Y when X = 0)

Purpose:

To predict the value of one variable based on another.

To understand the strength and direction of the relationship between two variables.

To find trends and make data-driven decisions.
'''
```

```python
#2 What are the key assumptions of Simple Linear Regression?
'''
Key Assumptions of Simple Linear Regression

Linearity:
The relationship between the independent variable (X) and the dependent variable (Y) should be linear.
‚Üí The data points should form a roughly straight-line pattern on a scatter plot.

Independence of Errors:
The residuals (errors between predicted and actual values) should be independent of each other.
‚Üí No correlation between consecutive error terms.

Homoscedasticity (Constant Variance):
The variance of the residuals should remain constant across all levels of X.
‚Üí The spread of errors should be roughly the same for small and large values of X.

Normality of Errors:
The residuals should follow a normal distribution.
‚Üí This is especially important for hypothesis testing and confidence intervals.

No (or minimal) Multicollinearity:
Since SLR has only one independent variable, this assumption is automatically satisfied.
(In multiple regression, it means independent variables should not be highly correlated.)
'''
```

```python
#3 Write the mathematical equation for a simple linear regression model and explain each term.
'''
Mathematical Equation of Simple Linear Regression

Y = Œ≤0 + Œ≤1X + Œµ

Explanation of Each Term:

Y (Dependent Variable):
The output or target value we are trying to predict (e.g., salary, house price).

X (Independent Variable):
The input or predictor variable used to make predictions (e.g., years of experience).

Œ≤‚ÇÄ (Beta Zero) ‚Äì Intercept:
The value of Y when X = 0.
It represents the point where the regression line crosses the Y-axis.

Œ≤‚ÇÅ (Beta One) ‚Äì Slope:
The change in Y for every one-unit increase in X.
A positive Œ≤‚ÇÅ means Y increases as X increases, and a negative Œ≤‚ÇÅ means Y decreases as X increases.

Œµ (Error Term / Residual):
The difference between the actual and predicted value of Y.
It accounts for randomness or factors not explained by X.
'''
```

```python
#5 What is the method of least squares in linear regression?
'''
Definition:
The method of least squares is a mathematical technique used to find the best-fitting line in linear regression.
It works by minimizing the sum of the squares of the errors (residuals) between the actual and predicted values of the dependent variable.

For each data point:

Error=(Actual Value ‚àí Predicted Value)

The goal is to minimize:

SSE = ‚àë(Yi‚àíYi‚Äã^‚Äã)2

Purpose:

To ensure the line of best fit has the smallest possible total error, providing the most accurate predictions for the given data.
'''
```

```python
#6 What is Logistic Regression? How does it differ from Linear Regression?
'''
Definition:
Logistic Regression is a supervised machine learning algorithm used for classification problems, where the output (dependent variable) is categorical,
such as Yes/No, Spam/Not Spam, or 0/1.

Instead of predicting exact numeric values, it predicts the probability that an observation belongs to a particular class.
It uses a logistic (sigmoid) function that converts any real number into a value between 0 and 1, which can be interpreted as a probability.
For example, it can predict whether an email is spam or not spam, or whether a patient has a disease or not.

Difference from Linear Regression

Logistic Regression differs from Linear Regression in its purpose and output.
Linear Regression predicts continuous values like house price or salary, whereas Logistic Regression predicts categorical outcomes like Yes/No or 0/1.
Linear Regression uses a straight-line equation 
 while Logistic Regression uses a sigmoid equation 


In Linear Regression, the output can be any real number, but in Logistic Regression, the output is always between 0 and 1, representing probability.
In short, Linear Regression is used for prediction, while Logistic Regression is used for classification.
'''
```

```python
#7  Name and briefly describe three common evaluation metrics for regression models.
'''
Three Common Evaluation Metrics for Regression Models

1.Mean Absolute Error (MAE):
MAE measures the average absolute difference between the actual and predicted values.
It shows how far predictions are from actual outcomes on average.
A lower MAE means better model performance.

MAE=n1 ‚àë‚à£Yi‚àíYi‚Äã^‚Äã‚à£

2.Mean Squared Error (MSE):
MSE calculates the average of the squared differences between actual and predicted values.
Squaring gives more weight to larger errors.
It helps identify models that make large prediction mistakes.

MSE=n1‚àë(Yi‚àíYi‚Äã^‚Äã)2

3.R-squared (R¬≤):
R¬≤ measures how well the regression model explains the variability in the dependent variable.
It ranges from 0 to 1, where values closer to 1 indicate a better fit.

R2=1‚àí SSt / SSr
```

```python
#8 What is the purpose of the R-squared metric in regression analysis?
'''
Purpose of the R-squared Metric in Regression Analysis

The R-squared (R¬≤) metric measures how well a regression model explains the variation in the dependent variable (Y) based on the independent variable(s) (X).
It shows the proportion of variance in the output that can be predicted from the input features.

R¬≤ values range from 0 to 1:

An R¬≤ of 1 means the model perfectly fits the data ‚Äî all predictions are accurate.

An R¬≤ of 0 means the model does not explain any of the variation in the target variable.

In simple terms, R-squared tells us how good our model is at predicting.
For example, an R¬≤ of 0.85 means that 85% of the variation in the dependent variable is explained by the model, while the remaining 15% is due to random noise or other unknown factors.
'''
```

```python
#9 Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

# Import necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Example data: Experience (X) vs Salary (Y)
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)   # Independent variable
Y = np.array([30000, 35000, 40000, 45000, 50000, 55000])  # Dependent variable

# Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Print slope and intercept
print("Slope (Œ≤1):", model.coef_[0])
print("Intercept (Œ≤0):", model.intercept_)
```
